```python
import pickle
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
DATASET_DIR = "modificado_result_final_resnet101.csv"
```

```python
df_resnet101 = pd.read_csv(DATASET_DIR)
labels_resnet101 = df_resnet101["Class"].to_numpy()
df_resnet101.drop(columns=["Class"], inplace=True)
df_resnet101.drop(columns=["Filename"], inplace=True)
df_resnet101_numpy = df_resnet101.to_numpy()
```

```python
scaler= StandardScaler()
X_resnet101 = scaler.fit_transform(df_resnet101_numpy)
```

```python
models_16=None
models_32=None
models_64=None
models_128=None
models_256=None
models_512=None

with open('ensemble_model_pca_16_resnet101.pkl', 'rb') as f:
    models_16 = pickle.load(f)

with open('ensemble_model_pca_32_resnet101.pkl', 'rb') as f:
    models_32 = pickle.load(f)

with open('ensemble_model_pca_64_resnet101.pkl', 'rb') as f:
    models_64 = pickle.load(f)

with open('ensemble_model_pca_128_resnet101.pkl', 'rb') as f:
    models_128 = pickle.load(f)

with open('ensemble_model_pca_256_resnet101.pkl', 'rb') as f:
    models_256 = pickle.load(f)

with open('ensemble_model_pca_512_resnet101.pkl', 'rb') as f:
    models_512 = pickle.load(f)
```

```python
x_resnet101_16 = models_16['pca'].transform(X_resnet101)
x_resnet101_32 = models_32['pca'].transform(X_resnet101)
x_resnet101_64 = models_64['pca'].transform(X_resnet101)
x_resnet101_128 = models_128['pca'].transform(X_resnet101)
x_resnet101_256 = models_256['pca'].transform(X_resnet101)
x_resnet101_512 = models_512['pca'].transform(X_resnet101)
```

```python
knn_16 = models_16['knn'].predict(x_resnet101_16)
knn_32 = models_32['knn'].predict(x_resnet101_32)
knn_64 = models_64['knn'].predict(x_resnet101_64)
knn_128 = models_128['knn'].predict(x_resnet101_128)
knn_256 = models_256['knn'].predict(x_resnet101_256)
knn_512 = models_512['knn'].predict(x_resnet101_512)
```

```python
print("KNN Results:")
print("PCA 16:", accuracy_score(labels_resnet101, knn_16))
print("PCA 32:", accuracy_score(labels_resnet101, knn_32))
print("PCA 64:", accuracy_score(labels_resnet101, knn_64))
print("PCA 128:", accuracy_score(labels_resnet101, knn_128))
print("PCA 256:", accuracy_score(labels_resnet101, knn_256))
print("PCA 256:", accuracy_score(labels_resnet101, knn_512))
```

```python
svm_16 = models_16['svm'].predict(x_resnet101_16)
svm_32 = models_32['svm'].predict(x_resnet101_32)
svm_64 = models_64['svm'].predict(x_resnet101_64)
svm_128 = models_128['svm'].predict(x_resnet101_128)
svm_256 = models_256['svm'].predict(x_resnet101_256)
svm_512 = models_512['svm'].predict(x_resnet101_512)
```

```python
print("SVM Results:")
print("PCA 16:", accuracy_score(labels_resnet101, svm_16))
print("PCA 32:", accuracy_score(labels_resnet101, svm_32))
print("PCA 64:", accuracy_score(labels_resnet101, svm_64))
print("PCA 128:", accuracy_score(labels_resnet101, svm_128))
print("PCA 256:", accuracy_score(labels_resnet101, svm_256))
print("PCA 256:", accuracy_score(labels_resnet101, svm_512))
```

```python
mlp_16 = models_16['mlp'].predict(x_resnet101_16)
mlp_32 = models_32['mlp'].predict(x_resnet101_32)
mlp_64 = models_64['mlp'].predict(x_resnet101_64)
mlp_128 = models_128['mlp'].predict(x_resnet101_128)
mlp_256 = models_256['mlp'].predict(x_resnet101_256)
mlp_512 = models_512['mlp'].predict(x_resnet101_512)
```

```python
print("MLP Results:")
print("PCA 16:", accuracy_score(labels_resnet101, mlp_16))
print("PCA 32:", accuracy_score(labels_resnet101, mlp_32))
print("PCA 64:", accuracy_score(labels_resnet101, mlp_64))
print("PCA 128:", accuracy_score(labels_resnet101, mlp_128))
print("PCA 256:", accuracy_score(labels_resnet101, mlp_256))
print("PCA 512:", accuracy_score(labels_resnet101, mlp_512))    
```

```python
rf_16 = models_16['rf'].predict(x_resnet101_16)
rf_32 = models_32['rf'].predict(x_resnet101_32)
rf_64 = models_64['rf'].predict(x_resnet101_64)
rf_128 = models_128['rf'].predict(x_resnet101_128)
rf_256 = models_256['rf'].predict(x_resnet101_256)
rf_512 = models_512['rf'].predict(x_resnet101_512)
```

```python
print("RF Results:")
print("PCA 16:", accuracy_score(labels_resnet101, rf_16))
print("PCA 32:", accuracy_score(labels_resnet101, rf_32))
print("PCA 64:", accuracy_score(labels_resnet101, rf_64))
print("PCA 128:", accuracy_score(labels_resnet101, rf_128))
print("PCA 256:", accuracy_score(labels_resnet101, rf_256))
print("PCA 512:", accuracy_score(labels_resnet101, rf_512))
```
